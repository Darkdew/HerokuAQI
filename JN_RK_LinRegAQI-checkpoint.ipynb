# Feature Importance

- For each feature in our model we can get its importance by using 'featur importance' property of the model
- it gives a score for each of the features, the higher the score, the higher is the feature's relevance to the label variable
- Feature Importance is an inbuilt class that comes with Tree-based-regressor
- If required we can use ExtraTreesRegressor for extracting the top 10 features of our datset.


```python
#implementing the featureimportance- SAMPLE
from sklearn.ensemble import ExtraTreesRegressor
model= ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(X.head())
print('\n')
print(model.feature_importances_)
```

```python
#plotting a bargraph per feature importances
feature_imp= pd.Series(model.feature_importances_, index=X.columns) #One-dimensional ndarray with axis labels 

f= feature_imp.nlargest(5)

f.plot(kind='barh')
plt.show()
```

# train_test_split

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state=0)
```

```python
LinRegModel= LinearRegression()
LinRegModel.fit(X_train,y_train)
```

```python
print("Coefficient of Deteremination/ R^2 on train set{}". format(LinRegModel.score(X_train, y_train)))
```

```python
print("Coefficient of Deteremination/ R^2 on test set{}". format(LinRegModel.score(X_test, y_test)))
```

```python
from sklearn.model_selection import cross_val_score

score= cross_val_score(LinRegModel, X, y, cv=5)

print(score)
```

- cross_val_sore divides the total data by the value provided as cv=''.
- so if total datapoints =100, it divides our data into 5 segments and then:uses the 1st segment as test and remaining 4 as train to calculate the R Square value
- in the next iteration: uses the 2nd segment as test and remaining 4 as train and thus iterates 5 times.
- It calculates R Square values for each of the 5 iterations

```python
score.mean()
```

```python
#creating a timer function

def timer(start_time= None):
    if not start_time:
        start_time= datetime.now()
        return start_time
    
    elif start_time:
        thour, temp_sec =divmod((datetime.now()-start_time).total_seconds(),3600)
        tmin,tsec= divmod(temp_sec, 60)
        print('time taken: {} hours {} minutes and {} seconds'.format(thour, tmin, round(tsec, 2)))
```

```python
start_time= timer(None) #timing to start here for start_time variable

grid_search.fit(X,y)

timer(start_time) #timing to stop here using the elif part of timer function
```
